```python
x = data.loc[:, ['X']]
y = data.loc[:, ['Y']]
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(9)
ax = plt.axes()
# ax.set_facecolor("#f0e6ef")   # 연핑크
ax.set_facecolor('#eafff5')     # 연초록
groups = data.groupby("Color")
for name, group in groups:
        plt.plot(group.X, group.Y, marker = "*", linestyle = '', color = name, markersize = 10)
        
# plt.show()

# 눈금 제거
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

plt.grid(True)

# 저장
plt.savefig('./christmas_tree.png')
```
